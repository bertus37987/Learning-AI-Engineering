### 3_Project Best Payed Data Jobs by the most important countrys

```python
"""Getting everything set up"""
# Imports (always)
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt

#Getting the Dataset
dataset_1 = load_dataset('lukebarousse/data_jobs')
# Transforming the Data into an Pandas df
df_1 = dataset_1['train'].to_pandas()
```

```python
#cleaning the Data
df_1 = df_1.dropna(subset=['salary_year_avg'])
#Creating a piviot table
country_salary_job_avg = df_1.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
#Extracting the top countrys
top_countrys = df_1['job_country'].value_counts().head(6).index
#Defining the jobs im personally looking for
jobs = ['Data Analyst','Machine Learning Engineer','Data Scientist']

#Changing the piviot to only include the countrys i want, after that, only the jobs i want
country_salary_job_avg = country_salary_job_avg.loc[top_countrys]
country_salary_job_avg = country_salary_job_avg[jobs]

#Plotting
country_salary_job_avg.plot(kind='bar')
plt.xticks(rotation=45,ha='right')
plt.xlabel("Job country")
plt.ylabel("Median Salary in $")
#Little Cleanup, so nothing gets covered
plt.legend(loc='upper right',bbox_to_anchor=(1.3, 1))
#No, its not AI, i just want to remove the annoying headline
plt.show()
```
