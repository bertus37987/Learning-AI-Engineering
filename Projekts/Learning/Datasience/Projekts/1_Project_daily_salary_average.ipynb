### 1_Project_daily_average_salary

```python
# Imports (always)
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#Getting the Dataset
dataset_1 = load_dataset('lukebarousse/data_jobs')
# Transforming the Data into an Pandas df
df_1 = dataset_1['train'].to_pandas()
```

```python
# Cleaning the Data 
df_1['job_posted_date'] = pd.to_datetime(df_1.job_posted_date)
df_1 = df_1.dropna(subset=['salary_year_avg'],how='all')

#Getting the job_posted_day
df_1['job_posted_day'] = df_1.job_posted_date.dt.day_of_year


daily_salary = df_1.groupby('job_posted_day')['salary_year_avg'].median().sort_index()

plt.title("Job salary average per day for 2023")
plt.xlabel("Day of the job post")# index !
plt.ylabel("Average salary") # values !
plt.plot(daily_salary.index, daily_salary.values)
```
