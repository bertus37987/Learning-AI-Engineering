### 2_Datasience_Pandas_DataAnalysis

```python
# Imports (always)
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#Getting the Dataset
dataset_1 = load_dataset('lukebarousse/data_jobs')
# Transforming the Data into an Pandas df
df_1 = dataset_1['train'].to_pandas()
```

```python
df_1['salary_year_avg'].median() # Get the median Salary
```

### Example: I want to see what is the lowest payed employee

```python
df_1['salary_year_avg'].min()
#But is this Really helpfull, no !
df_1['salary_year_avg'].idxmin()
#This returns the Index
df_1.iloc[665729]
#Now this is helpfull
```

### A few other usefull methods

```python
df_1['job_title_short'].value_counts()
```

# Grouping Data (very important)

```python
# Groupes the Data by index= job_title_short and job_country, the values are the median of salary_year/hour_avg
df_1.groupby(['job_title_short','job_country'])[['salary_year_avg','salary_hour_avg']].median()
```

```python
# Now this is a bit more advanct, but when looked at step by step its just everything added together, if you want to pass multiple arguments you put .agg
df_1.groupby(['job_title_short'])[['salary_year_avg','salary_hour_avg']].agg(['median','max','min','count']).sort_values(by=('salary_year_avg','median'))
```
